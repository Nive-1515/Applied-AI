```python
import numpy as np

# Step activation function
def step_function(net):
    return 1 if net >= 0 else 0

# Perceptron training function
def train_perceptron(X, y, w, b, lr):
    converged = False
    epoch = 0

    while not converged:
        converged = True
        epoch += 1
        print(f"\nEpoch {epoch}")

        for i in range(len(X)):
            # Net input
            net = np.dot(w, X[i]) + b
            output = step_function(net)

            print(f"Input: {X[i]}, Net: {net:.2f}, Output: {output}, Target: {y[i]}")

            # Weight and bias update if misclassified
            if output != y[i]:
                error = y[i] - output
                w = w + lr * error * X[i]
                b = b + lr * error
                converged = False

                print(f"Updated Weights: {w}, Updated Bias: {b:.2f}")

    return w, b

# Input patterns
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

learning_rate = 0.1

# -------- OR GATE --------
print("\nTraining OR Gate")
y_or = np.array([0, 1, 1, 1])
w_or = np.array([0.2, 0.2])
b_or = -0.1

final_w_or, final_b_or = train_perceptron(X, y_or, w_or, b_or, learning_rate)

print("\nFinal OR Gate Weights:", final_w_or)
print("Final OR Gate Bias:", final_b_or)

# -------- AND GATE --------
print("\nTraining AND Gate")
y_and = np.array([0, 0, 0, 1])
w_and = np.array([0.2, 0.2])
b_and = -0.3

final_w_and, final_b_and = train_perceptron(X, y_and, w_and, b_and, learning_rate)

print("\nFinal AND Gate Weights:", final_w_and)
print("Final AND Gate Bias:", final_b_and)
```

```python
import numpy as np

# Step activation function
def step_function(net):
    return 1 if net >= 0 else 0

# Perceptron training with fixed epochs
def train_perceptron_fixed_epochs(X, y, w, b, lr, epochs):
    for epoch in range(1, epochs + 1):
        print(f"\nEpoch {epoch}")
        for i in range(len(X)):
            net = np.dot(w, X[i]) + b
            output = step_function(net)

            print(f"Input: {X[i]}, Net: {net:.2f}, Output: {output}, Target: {y[i]}")

            # Update rule
            error = y[i] - output
            if error != 0:
                w = w + lr * error * X[i]
                b = b + lr * error
                print(f"Updated Weights: {w}, Updated Bias: {b:.2f}")

    return w, b

# Input patterns
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

learning_rate = 0.1
epochs = 3

# -------- OR Gate --------
print("\nTraining OR Gate")
y_or = np.array([0, 1, 1, 1])
w_or = np.array([0.2, 0.2])
b_or = -0.1

final_w_or, final_b_or = train_perceptron_fixed_epochs(
    X, y_or, w_or, b_or, learning_rate, epochs
)

print("\nFinal OR Gate Weights:", final_w_or)
print("Final OR Gate Bias:", final_b_or)

# -------- AND Gate --------
print("\nTraining AND Gate")
y_and = np.array([0, 0, 0, 1])
w_and = np.array([0.2, 0.2])
b_and = -0.3

final_w_and, final_b_and = train_perceptron_fixed_epochs(
    X, y_and, w_and, b_and, learning_rate, epochs
)

print("\nFinal AND Gate Weights:", final_w_and)
print("Final AND Gate Bias:", final_b_and)
```
