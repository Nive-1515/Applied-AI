```python
import cv2
import numpy as np
import imutils
import time
from matplotlib import pyplot as plt

# Load image
image_path = r"/content/817n0Ovt5NL._AC_UF894,1000_QL80_.jpg"
image = cv2.imread(image_path)

if image is None:
    print("Error: Image not found!")
    exit()

orig = image.copy()

# Convert to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()

# Convert to grayscale and blur
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (7, 7), 0)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(blur, cmap='gray')
plt.title("Blurred")
plt.axis("off")
plt.show()

# Edge detection
edged = cv2.Canny(blur, 50, 100)

plt.figure(figsize=(6,6))
plt.imshow(edged, cmap='gray')
plt.title("Edges")
plt.axis("off")
plt.show()

# Morphological operations
edged = cv2.dilate(edged, None, iterations=1)
edged = cv2.erode(edged, None, iterations=1)

plt.figure(figsize=(6,6))
plt.imshow(edged, cmap='gray')
plt.title("Processed Edges")
plt.axis("off")
plt.show()

# Thresholding
_, thresh = cv2.threshold(blur, 0, 255,
                         cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(6,6))
plt.imshow(thresh, cmap='gray')
plt.title("Threshold")
plt.axis("off")
plt.show()

# Find contours
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL,
                        cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)

object_count = 0
used_positions = []

start_time = time.time()

for c in cnts:
    if cv2.contourArea(c) < 50:
        continue

    object_count += 1

    rect = cv2.minAreaRect(c)
    box = cv2.boxPoints(rect)
    box = np.intp(box)

    # Draw bounding box
    cv2.drawContours(image, [box], -1, (0, 255, 0), 2)

    w, h = rect[1]
    x, y, bw, bh = cv2.boundingRect(c)

    y_text = max(y - 10, 20)

    # Avoid overlapping labels
    for (px, py) in used_positions:
        if abs(x - px) < 80 and abs(y_text - py) < 30:
            y_text += 25

    used_positions.append((x, y_text))

    label = f"{int(w)}x{int(h)}"

    cv2.rectangle(image, (x, y_text - 15),
                  (x + 100, y_text + 5),
                  (0, 255, 0), -1)

    cv2.putText(image, label, (x + 2, y_text),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                (0, 0, 0), 1)

end_time = time.time()
latency = end_time - start_time

# Display count & latency
cv2.putText(image, f"Count: {object_count}",
            (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.8,
            (255, 0, 0), 2)

cv2.putText(image, f"Latency: {latency:.3f}s",
            (10, 60),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6,
            (0, 0, 255), 2)

# Final output
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.imshow(image_rgb)
plt.title(f"Final Output: {object_count} Shapes")
plt.axis("off")
plt.show()

print(f"Total shapes detected: {object_count}")
print(f"Processing latency: {latency:.3f} seconds")
```
