```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read image
image = cv2.imread('/content/images (2).jpeg')   # change filename if needed
orig = image.copy()

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Blur to reduce noise
blur = cv2.GaussianBlur(gray, (11,11), 0)

# Threshold (better than edges for objects like coins)
_, thresh = cv2.threshold(blur, 120, 255, cv2.THRESH_BINARY_INV)

# Remove noise
kernel = np.ones((3,3), np.uint8)
thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

valid_contours = []

for cnt in contours:
    area = cv2.contourArea(cnt)

    # Filter small noise
    if area > 300:
        valid_contours.append(cnt)

# Count objects
print("Objects detected:", len(valid_contours))

# 🔷 Measurement (IMPORTANT)
# Assume 1 coin width = 2.5 cm (reference)
pixels_per_cm = None

# Find reference object (largest contour)
largest = max(valid_contours, key=cv2.contourArea)
x,y,w,h = cv2.boundingRect(largest)

# Assume largest object width = 2.5 cm
pixels_per_cm = w / 2.5

# Draw and measure all objects
for cnt in valid_contours:
    x,y,w,h = cv2.boundingRect(cnt)

    width_cm = w / pixels_per_cm
    height_cm = h / pixels_per_cm

    cv2.rectangle(orig,(x,y),(x+w,y+h),(0,255,0),2)
    cv2.putText(orig,f"{width_cm:.2f}cm",
                (x,y-10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.6,(0,255,0),2)

# Show result
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
plt.title("Object Counting & Measurement")
plt.axis("off")
plt.show()
```
