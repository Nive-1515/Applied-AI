```python
# Install libraries
!pip install opencv-python imutils

# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from imutils.object_detection import non_max_suppression

# -------------------------------
# 1. Upload Image
# -------------------------------
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()

# -------------------------------
# 2. Convert to Grayscale
# -------------------------------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Improve contrast (VERY IMPORTANT)
gray = cv2.equalizeHist(gray)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale + Equalized Image")
plt.axis('off')
plt.show()

# -------------------------------
# 3. Load Viola–Jones Classifier
# -------------------------------
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# -------------------------------
# 4. Detect Faces (Tuned)
# -------------------------------
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.3,     # controls scaling
    minNeighbors=6,      # increase → remove false positives
    minSize=(40, 40)     # ignore small fake detections
)

print("Before NMS:", len(faces))

# -------------------------------
# 5. Remove Duplicate Boxes (NMS)
# -------------------------------
if len(faces) > 0:
    faces = non_max_suppression(np.array(faces), overlapThresh=0.3)

# -------------------------------
# 6. Draw Final Boxes
# -------------------------------
output = img_rgb.copy()

for (x, y, w, h) in faces:
    cv2.rectangle(output, (x, y), (x+w, y+h), (0, 255, 0), 2)

# -------------------------------
# 7. Show Final Output
# -------------------------------
plt.imshow(output)
plt.title(f"Faces detected: {len(faces)}")
plt.axis('off')
plt.show()

print("Final number of faces detected:", len(faces))
```
