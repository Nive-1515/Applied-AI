```python
import tensorflow as tf
from tensorflow.keras import layers, models
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ------------------ Build CNN Model ------------------
def build_model():
    inputs = tf.keras.Input(shape=(128, 128, 3))

    x = layers.Conv2D(32, (3, 3), activation='relu')(inputs)
    x = layers.MaxPooling2D(2, 2)(x)

    x = layers.Conv2D(64, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D(2, 2)(x)

    x = layers.Conv2D(128, (3, 3), activation='relu')(x)
    x = layers.MaxPooling2D(2, 2)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    outputs = layers.Dense(1, activation='sigmoid')(x)

    return models.Model(inputs, outputs)

model = build_model()
print("✅ CNN Model Built Successfully!")

# ------------------ Load Image ------------------
image_path = "/content/images (3).jpeg"   # 🔁 PUT YOUR IMAGE PATH HERE
img = cv2.imread(image_path)

if img is None:
    raise ValueError("❌ Image not found! Check path")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ------------------ Preprocessing ------------------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Threshold
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Sobel Edge Detection
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)

feature_map = cv2.magnitude(sobelx, sobely)
feature_map = cv2.normalize(feature_map, None, 0, 255, cv2.NORM_MINMAX)
feature_map = np.uint8(feature_map)

# ------------------ Face Detection (Haar Cascade) ------------------
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3,minSize=(30,30))

img_faces = img_rgb.copy()

for (x, y, w, h) in faces:
    cv2.rectangle(img_faces, (x, y), (x + w, y + h), (0, 255, 0), 2)

print(f"✅ Faces detected: {len(faces)}")

# ------------------ Display Results ------------------
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(thresh, cmap='gray')
plt.title("Threshold Image")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(feature_map, cmap='gray')
plt.title("Feature Map (Sobel)")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(img_faces)
plt.title(f"Final Detection - Faces: {len(faces)}")
plt.axis("off")

plt.tight_layout()
plt.show()
```
