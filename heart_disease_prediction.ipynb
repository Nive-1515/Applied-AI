```python
import numpy as np
```

```python
import pandas as pd

df = pd.read_csv("heart.csv")
df.head()
```

```python
X = df.drop(['age', 'sex', 'trtbps', 'output'], axis=1).values
y = df.output.values
```

```python
print(X)
print(X.shape)
```

```python
print(y)
print(y.shape)
```

```python
unique_values = np.unique(y)
counts = np.zeros_like(unique_values)

for i, value in enumerate(unique_values):
  counts[i] = np.count_nonzero(y == value)

for i, value in enumerate(unique_values):
  print(f"Class label: {value}, Count: {counts[i]}")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.2, random_state=0)
```

```python
print(X_train)
print(X_train.shape)
```

```python
print(X_test)
print(X_test.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
print(y_test)
print(y_test.shape)
```

```python
unique_values = np.unique(y_test)
counts = np.zeros_like(unique_values)

for i, value in enumerate(unique_values):
  counts[i] = np.count_nonzero(y_test == value)

for i, value in enumerate(unique_values):
  print(f"Class label: {value}, Count: {counts[i]}")
```

```python
import warnings
warnings.filterwarnings("ignore")
from collections import Counter
```

```python
from imblearn.over_sampling import SMOTE, SVMSMOTE, ADASYN
```

```python
counter = Counter(y_train)
print('Before',counter)
# oversampling the train dataset using SMOTE + ENN
smenn = SVMSMOTE()
X_train_svmsm, y_train_svmsm = smenn.fit_resample(X_train, y_train)

counter = Counter(y_train_svmsm)
print('After',counter)
```

```python
import sklearn
from sklearn import metrics
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import precision_recall_curve,roc_curve, auc, roc_auc_score
from sklearn.metrics import balanced_accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train_svmsm, y_train_svmsm.ravel())
```

```python
preds = logistic_model.predict(X_test)
```

```python
accuracy_score(y_test, preds)
```

```python
classes = ["Normal","AbNormal"]
cm = confusion_matrix(y_test,preds)

print('Confusion Matrix')
print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                display_labels=classes)
fig, ax = plt.subplots(figsize=(5,5))
plt.title("Confusion Matrix")
disp = disp.plot(ax=ax)
plt.show()
```

```python
Accuracy = metrics.accuracy_score(y_test, preds)
print('Accuracy:', Accuracy*100)

Precision = metrics.precision_score(y_test, preds)
print('Precision:', Precision*100)

Recall = metrics.recall_score(y_test, preds)
print('Recall:', Recall*100)

F1_score = metrics.f1_score(y_test, preds)
print('F1 Score:', F1_score*100)
```

```python
roc_auc = roc_auc_score(y_test,preds)
print('ROC AUC score: %.3f' % roc_auc)
print('Mean ROC AUC: %.5f' % roc_auc.mean())

fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test,preds)

auc_keras = auc(fpr_keras, tpr_keras)
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--' ,'No Skill')
plt.plot(fpr_keras, tpr_keras, label = 'area = {:.3f}'.format(auc_keras))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc = 'best')
plt.show()
```

```python
target_names=['Normal', 'Abnormal']
print(classification_report(preds, y_test,target_names=target_names))
```
